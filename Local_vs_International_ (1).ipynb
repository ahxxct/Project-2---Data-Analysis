**Descriptive Statistics**

```python
cols = ['shopping_frequency_num','brand_preference_num','main_factor_num']

for i in cols:
    print("-----", i, "-----")
    print("Mean:", df[i].mean())
    print("Median:", df[i].median())
    print("Mode:", df[i].mode().values)
```

**Gender Distribution**

```python
G = df['Gender'].value_counts()

# 2. Create the Pie Chart
# autopct='%1.1f%%' displays the percentages on the slices
plt.pie(G, labels=G.index, autopct='%1.1f%%', colors=['pink', 'lightblue'], startangle=90)

plt.title('Proportion of Male vs. Female')

plt.show()
```

The chart shows the distribution of participants by gender, with a relatively balanced representation of males and females.
This balance helps ensure that the results are not biased toward a specific gender group.

**Shopping Frequency**

```python
frequency_counts = df['Shopping_Frequency'].value_counts()

plt.bar(frequency_counts.index, frequency_counts.values, color=['mediumpurple', 'skyblue', 'lightgreen', 'gray'])

plt.title('Distribution of Shopping Frequency')
plt.xlabel('Shopping Frequency')
plt.ylabel('Number of Respondents')

plt.yticks(range(0, max(frequency_counts.values) + 1))
plt.show()
```

The chart shows the distribution of shopping frequency among participants, with most respondents purchasing clothing either monthly or every season.
This indicates that clothing shopping is a regular habit for many consumers rather than a rare activity.

Brand **Preference**

```python
brand_counts = df['Brand_Preference'].value_counts()

plt.pie(brand_counts.values, labels=brand_counts.index, autopct='%1.1f%%', colors=['skyblue', 'mediumpurple', 'lightgreen'])

plt.title('Brand Preference Distribution')

plt.show()
```

The chart shows the distribution of brand preferences among participants, with a large proportion favoring international brands or both local and international options.

This indicates that while international brands are popular, local brands still play an important role in consumer choices.

The high percentage of participants choosing both equally suggests that consumers do not rely on a single type of brand and tend to balance between local and international options.

**Shopping Method**

```python
method_counts = df['Shopping_Method'].value_counts()

plt.pie(method_counts.values, labels=method_counts.index, autopct='%1.1f%%', colors=['skyblue', 'mediumpurple', 'lightgreen'])

plt.title('Preferred Shopping Method')
plt.show()
```

The chart shows that a large proportion of participants prefer a combination of online and in-store shopping methods.
This indicates that consumers value both convenience and the ability to physically inspect products before purchasing.

**Main Factor**

```python
factor_counts = df['Main_Factor'].value_counts()

plt.bar(factor_counts.index, factor_counts.values, color=['mediumpurple', 'skyblue', 'lightgreen', 'gray'])

plt.title('Most Important Factor When Choosing Clothing')
plt.xlabel('Factor')
plt.ylabel('Number of Respondents')

plt.yticks(range(0, max(factor_counts.values) + 1))

plt.show()
```

The chart shows that price and quality are the most important factors influencing consumer decisions.
This indicates that consumers prioritize value for money over brand name when choosing clothing.

A high preference for design suggests that style and appearance play a significant role in purchasing decisions.

**Local Price**

```python
price_counts = df['Local_Price'].value_counts()

plt.pie(price_counts.values, labels=price_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'mediumpurple', 'skyblue'])

plt.title('Price Comparison of Local vs International Brands')

plt.show()
```

The chart shows that a large proportion of participants perceive local brands as cheaper compared to international brands.
This indicates that affordability is a key factor that makes local brands more attractive to consumers.

**International Worth (Worth the price?)**

```python
worth_counts = df['Intl_Worth'].value_counts()

plt.pie(worth_counts.values, labels=worth_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])

plt.title('Are International Brands Worth Their Price?')

plt.show()
```

The chart shows that participants are evenly divided in their opinions on whether international brands are worth their price.
This indicates that consumers have mixed perceptions, with no clear consensus regarding the value of international brands.

**Reasons for Choosing Local Brands**

```python
local_reasons = df.filter(like='local').sum().sort_values()

plt.barh(local_reasons.index, local_reasons.values)

plt.title('Reasons for Choosing Local Brands')
plt.xlabel('Number of Responses')
plt.ylabel('Reasons')

plt.show()
```

The chart shows that better quality and lower price are the most important reasons for choosing local brands among participants.
This indicates that consumers are primarily driven by value for money when selecting local products. Additionally, supporting local businesses and better fit also play a significant role in influencing their choices.

**Reasons for Choosing International Brands**

```python
intl_reasons = df.filter(like='intl').sum().sort_values()

plt.barh(intl_reasons.index, intl_reasons.values)

plt.title('Reasons for Choosing International Brands')
plt.xlabel('Number of Responses')
plt.ylabel('Reasons')

plt.tight_layout()
plt.show()
```

The chart shows that better quality and trust are the most important reasons for choosing international brands among participants.
This indicates that consumers associate international brands with higher reliability and superior performance. Additionally, factors such as better fit and brand name also contribute to their preference.
